```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

import os
```

```python
def boot_strap(arr: np.array,
               sample_size: int,
              loop_size: int):
    
    output = []
    
    plt.figure(figsize=(10, 5))
    for item in list(range(0, loop_size)):
        _ = np.random.choice(arr, sample_size)
        sns.kdeplot(_)
        output.append(_.mean())
    
    plt.figure(figsize=(10, 5))
    sns.kdeplot(output)
    return output
```

```python
if __name__ == "__main__":
#     data = pd.read_csv(os.path.join(DATA_DIR, "data.csv"))
    data = np.random.randint(200, size=10000)
    
    print(data.mean())
    
    _ = boot_strap(data,
          sample_size=500,
          loop_size=20)
    
    print(pd.DataFrame(_))
```
